# Python Problem Solving

So you're given a complicated problem, and you don't know where to start.


1. Understand the Problem
    * Ex. my interview, difference between a palindrome/anagram
    * Do you need more domain knowledge? Modeling physics problems?
2. Work Example by Hand
    * Solve small/simple instance
    * If the problem has many steps, solve one at a time
    * Generally try 3 examples
3. Write down what you did
    * Write down the exact steps for just that instance in “pseudocode” or comments
        * We as humans have common sense, and often skip steps
        * Write down everything, eli5
4. Find Patterns(optimize)
    * We want an efficient algo for most/all instances of our problem
    * Are we repeating ourselves? - can be loops
    * Are there conditions? if/else/then - use conditionals
    * Values - using same over and over? Using different each time? Carry over from other steps
    * Allows us to write less/ more understandable code
    * If this is difficult try a different example for steps 1 and 2
5. Check By hand
    * Follow the path for simple example
    * Check for a few inputs
6. Translate into code
    * Choose your programming language (if you have multiple to choose from)
    * Check as you go with print statements, with simple inputs that you had as examples, know what to expect
7. Run Test Case
    * Did we get the right answer?
    * if , yes try a few more ex until confident that it works. Try to think of edge cases
    * If no, debug
8. Debug Failed tests
    * If algo path wrong, go back to step 3
    * Using print statements find where things went wrong
        * Follow the expected answers until you find something you didn’t expect

```
take two inputs word1 and word2
if the length of word1 not equal to the length of word2
    not an anagram
for each letter in word1
    if the letter is in word2
        take out the letter
    elif the letter is not in word2
        word2 is not an anagram

we've checked all letters in word 1, and there are no mismatches with word2, then it's an anagram!
```

#### Check By Hand

Run through one of our examples where we already know the answer

#### Translate what we have now into code
No peaking! We'll do this in class! Will post after covered in class

#### Run Test Cases

#### Debug Failed tests


#### Is it good enough?... Or can we do better?
(potential step 9)
